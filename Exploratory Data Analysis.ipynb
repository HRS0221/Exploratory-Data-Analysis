# 2-D Scater Plot :)

```python
iris.plot(kind = 'scatter', x = 'SepalLengthCm', y = 'SepalWidthCm');
plt.show()
iris.plot(kind = 'scatter', x = 'PetalLengthCm', y = 'PetalWidthCm');
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(iris, hue = "Species") \
    .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
    .add_legend();
plt.savefig("Sepal.png");
plt.show();
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(iris, hue = "Species") \
    .map(plt.scatter, "PetalLengthCm", "PetalWidthCm") \
    .add_legend();
plt.savefig("Petal.png");
plt.show();
```

# Pairplot Plot :)

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(iris, hue = "Species", height = 3, diag_kind= "kde");
plt.figtext(0.4, 0.01, 'Author: Himanshu Salunke', ha='center', va='center', fontsize=10, color='gray')
plt.savefig("Pairplot.png")
plt.show()
```
